```python
from __future__ import division
from math import sqrt,log
import numpy as np
```

```python
X = np.random.exponential(4,100)+np.random.chisquare(4,100)+np.random.normal(3,41,100)
```

```python
n = len(X)
def Fn(x):
    return((1/n)*sum([1*(X[i]<=x) for i in range(n)]))
def U(x,a):
    e = sqrt((log(2/a))/(2*n))
    return(min(1,Fn(x)+e))
def L(x,a):
    e = sqrt((log(2/a))/(2*n))
    return(max(Fn(x)-e,0))
```

```python
import matplotlib.pylab as plt 
from math import e
%matplotlib inline
```

```python
p = plt.figure(figsize=(12,6))
p = plt.xlabel('Eje X')
p = plt.ylabel('$F_{n}$')
p = plt.plot(np.linspace(min(X),max(X),50),
        [Fn(x) for x in np.linspace(min(X),max(X),50)],
            color='red',
            lw=3,
            label = 'Funcion de Distribucion Empirica')
p = plt.plot(np.linspace(min(X),max(X),50),
        [U(x,0.1) for x in np.linspace(min(X),max(X),50)],
            color='blue',
            lw=3,
            label = '$U(x)$')
p = plt.plot(np.linspace(min(X),max(X),50),
        [L(x,0.1) for x in np.linspace(min(X),max(X),50)],
            color='green',
            lw=3,
            label = '$L(x)$')
p= plt.legend()
```

```python
def Q(p) :
    return(min([x for x in X if Fn(x)>p]))
```
