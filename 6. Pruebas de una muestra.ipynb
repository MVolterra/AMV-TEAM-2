```python
import numpy as np
import matplotlib.pylab as plt
from math import exp,log,sqrt,pi,sin
import scipy.stats as sts
import pandas as pd
%matplotlib inline
```

```python
n = 10
tabla = pd.DataFrame(columns=['X','Y'],index=range(1,n+1))
tabla['X'] = np.random.uniform(0,1,n)
tabla['Y'] = tabla['X']+np.random.normal(0,0.1,n)
tabla['$|Z|$'] = abs(tabla['X'] - tabla['Y'])
tabla = tabla.sort_values(['$|Z|$'])
tabla['R'] = range(1,n+1)
tabla = tabla.sort_index()
tabla['$\\varphi$'] = 1*((tabla['X'] - tabla['Y'])>0)
tabla['$R\\varphi$'] = tabla['R']*tabla['$\\varphi$']
T = sum(tabla['$R\\varphi$'])
print(T)
tabla
```
